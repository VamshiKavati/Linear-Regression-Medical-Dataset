# Data preprocessing

```python
#Dummy variable
Categorical_columns = ['sex', 'children', 'smoker', 'region']
df_encode = pd.get_dummies(data = df , prefix = 'OHE', prefix_sep= '_', columns = Categorical_columns, drop_first = True,
                          dtype = 'int8')
df_encode.head()
```

```python
print('Columns in orginal data Frame: ', df.columns.values)
print('\nNumber of rows and columns in the dataset:', df.shape)
print('\n Columns in data frame after encoding dummy variables: ', df_encode.columns.values)
print('\n Number of rows and columns in the dataset : ', df_encode.shape)
```

```python
from scipy.stats import boxcox
y_bc, lam, ci = boxcox(df_encode['charges'], alpha = 0.05)

ci , lam , y_bc
```

```python
df_encode['charges'] = np.log(df_encode['charges'])
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x = df_encode.drop('charges', axis = 1)
y = df_encode['charges']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 23)
```

# Model Building

```python
#step 1: add x0 = 1 to dataset
x_train_0 = np.c_[np.ones((x_train.shape[0], 1)), x_train]
x_test_0 = np.c_[np.ones((x_test.shape[0], 1)), x_test]

#step 2: build model
theta = np.matmul(np.linalg.inv(np.matmul(x_train_0.T, x_train_0)), np.matmul(x_train_0.T, y_train))
theta
```

```python
parameter = ['theta_'+str(i) for i in range(x_train_0.shape[1])]
columns = ['intersect: X_0 = 1'] + list(x.columns.values)
parameter_df = pd.DataFrame({'Parameter': parameter, 'Column': columns, 'theta':theta})
parameter_df
```

```python
#scikit Learn Model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train )

#Parameter
sk_theta = [lin_reg.intercept_]+list(lin_reg.coef_)
parameter_df = parameter_df.join(pd.Series(sk_theta, name='Sklearn_theta'))
parameter_df
```

# Model Evaluation

```python
#Normal equation
y_pre_n = np.matmul(x_test_0, theta)

#Evaluation : MSE
j_mse = np.sum((y_pre_n - y_test)**2) / x_test_0.shape[0]

#R square
sse = np.sum((y_pre_n - y_test) ** 2)
sst = np.sum((y_test - y_test.mean())**2)

r2 = 1 - (sse/sst)
print('Mean Squared error: ', j_mse)
print('R - Squared : ', r2)
```

```python
#sklearn Regression model
y_pred = lin_reg.predict(x_test)

#evaluate mse
from sklearn.metrics import mean_squared_error
J_mse = mean_squared_error(y_pred, y_test)

#r2
R2 = lin_reg.score(x_test, y_test)
print('Mean Squared error: ', J_mse)
print('R - Squared : ', R2)
```

# Model Validation

```python
#Check of Linearity
f = plt.figure(figsize = (14, 5))
ax = f.add_subplot(121)
sns.scatterplot(y_test, y_pred, ax=ax, color='r')
ax.set_title('Check for Linearity:\n Actual vs Predicted value')

#check for residual normality and mean
ax = f.add_subplot(122)
sns.distplot((y_test - y_pred), ax = ax , color = 'b')
ax.axvline((y_test - y_pred).mean(), color = 'k', linestyle = "--")
ax.set_title('Check for Residual normality & mean:\n Residual error')
plt.show()
```

```python
#check for multivariate Normality
#Quantile - Quantile Plot

f, ax = plt.subplots(1, 2, figsize = (14, 6))
import scipy as sp
_,(_,_,r)= sp.stats.probplot((y_test - y_pred),fit=True,plot=ax[0])
ax[0].set_title('Check for Multivariate Normality: \nQ-Q Plot')

#Check for Homoscedasticity
sns.scatterplot(y = (y_test -y_pred), x = y_pred, ax=ax[1], color = 'r')
ax[1].set_title('Check for HomoScedasticity:\nResidual Vs Redicted');
plt.show()
```
